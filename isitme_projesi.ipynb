```python
import cv2
import os
import json

def extract_frames(video_path, output_folder, output_json, fps=30):
    """Videoyu frame'lere ayırır ve JSON dosyasına kaydeder."""
    
    # Klasör oluştur
    os.makedirs(output_folder, exist_ok=True)
    
    # Video aç
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    frame_data = {"frames": []}
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Video bitti

        # Frame kaydet
        frame_filename = f"frame_{frame_count:04d}.jpg"
        frame_path = os.path.join(output_folder, frame_filename)
        cv2.imwrite(frame_path, frame)

        # Zaman hesaplama
        timestamp = frame_count / fps

        # JSON içine frame bilgisi ekleme
        frame_info = {
            "frame_id": frame_count + 1,
            "image_path": frame_path,
            "timestamp": round(timestamp, 3),
            "keypoints": []  # İskelet verisi varsa buraya eklenebilir
        }
        frame_data["frames"].append(frame_info)
        
        frame_count += 1

    cap.release()

    # JSON dosyasını kaydet
    with open(output_json, "w", encoding="utf-8") as f:
        json.dump(frame_data, f, indent=4, ensure_ascii=False)

    print(f"{frame_count} frame çıkarıldı ve JSON kaydedildi: {output_json}")

# Kullanım
video_path = "İşaretDiliVideo\Ben.mp4"  # İşlenecek video
output_folder = "frames"  # Frame'lerin kaydedileceği klasör
output_json = "frames_data_ben.json"  # Çıktı JSON dosyası

extract_frames(video_path, output_folder, output_json)
```

```python
import speech_recognition as sr
import json
import cv2
import os
import time

# JSON dosyalarının bulunduğu klasör
json_klasor = "jsonDosyası"

# Tüm JSON dosyalarını okuyarak veriyi yükle
frame_data = {}

for dosya in os.listdir(json_klasor):
    if dosya.endswith(".json"):
        kelime = dosya.replace(".json", "")  # Dosya adından kelimeyi al (örneğin: "merhaba.json" → "merhaba")
        with open(os.path.join(json_klasor, dosya), "r", encoding="utf-8") as f:
            json_verisi = json.load(f)
            frame_data[kelime] = [frame["image_path"] for frame in json_verisi.get("frames", [])]  # Tüm frame path'leri al

# Ses tanıma fonksiyonu
def sesli_komut_al():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Lütfen bir şeyler söyleyin...")
        audio = r.listen(source)

        try:
            metin = r.recognize_google(audio, language="tr-TR")
            print(f"Duyduğum: {metin}")
            return metin.lower()
        except sr.UnknownValueError:
            print("Sesi anlayamadım.")
            return ""
        except sr.RequestError as e:
            print(f"Google API hatası: {e}")
            return ""

# Frame'leri oynatan fonksiyon
def animasyon_oynat(kelime):
    if kelime in frame_data:
        print(f"'{kelime}' kelimesi için animasyon oynatılıyor...")
        cv2.namedWindow("İşaret Dili Animasyonu", cv2.WINDOW_NORMAL)
        cv2.resizeWindow("İşaret Dili Animasyonu", 640, 480)

        for frame_path in frame_data[kelime]:  # İlgili kelimenin frame'lerini sırayla aç
            frame = cv2.imread(frame_path)
            if frame is not None:
                cv2.imshow("İşaret Dili Animasyonu", frame)
                cv2.waitKey(100)  # 100ms bekleyerek animasyon efekti ver
                time.sleep(0.1)  # Gözle görülür animasyon için ek gecikme
            else:
                print(f"Frame okunamadı: {frame_path}")

        cv2.waitKey(500)  # Son kare ekranda biraz daha uzun kalsın
        cv2.destroyAllWindows()
    else:
        print(f"'{kelime}' için animasyon bulunamadı.")

# Ana akış
metin = sesli_komut_al()
if metin:
    kelimeler_in_metin = metin.split()
    for kelime in kelimeler_in_metin:
        animasyon_oynat(kelime)
```

```python
import speech_recognition as sr
import json
import pygame
import os

# Pygame'i başlat
pygame.init()

# Ekran boyutları
SCREEN_WIDTH = 640
SCREEN_HEIGHT = 480
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("İşaret Dili Animasyonu")

# JSON dosyalarının bulunduğu klasör
json_klasor = "jsonDosyası"  # Burada video frame'lerini içeren JSON dosyaları olmalı

# Tüm JSON dosyalarını okuyarak veriyi yükle
frame_data = {}

for dosya in os.listdir(json_klasor):
    if dosya.endswith(".json"):
        kelime = dosya.replace(".json", "")  # Dosya adından kelimeyi al (örneğin: "merhaba.json" → "merhaba")
        with open(os.path.join(json_klasor, dosya), "r", encoding="utf-8") as f:
            json_verisi = json.load(f)
            frame_data[kelime] = [frame["image_path"] for frame in json_verisi.get("frames", [])]  # Tüm frame path'lerini al

# Ses tanıma fonksiyonu
def sesli_komut_al():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Lütfen bir şeyler söyleyin...")
        audio = r.listen(source)

        try:
            metin = r.recognize_google(audio, language="tr-TR")
            print(f"Duyduğum: {metin}")
            return metin.lower()  # Küçük harfe çeviriyoruz, karşılaştırmalar için
        except sr.UnknownValueError:
            print("Sesi anlayamadım.")
            return ""
        except sr.RequestError as e:
            print(f"Google API hatası: {e}")
            return ""

# Frame'leri oynatan fonksiyon
def animasyon_oynat(kelime):
    # Kelimeyi frame_data'da arıyoruz
    if kelime in frame_data:
        print(f"'{kelime}' kelimesi için animasyon oynatılıyor...")

        # Frame'leri belleğe al
        frames = [pygame.image.load(frame_path) for frame_path in frame_data[kelime] if os.path.exists(frame_path)]

        if not frames:
            print(f"'{kelime}' için hiç geçerli frame bulunamadı.")
            return

        # Animasyonu oynat
        for frame in frames:
            screen.fill((0, 0, 0))  # Ekranı temizle
            screen.blit(frame, (0, 0))  # Frame'i ekrana yerleştir
            pygame.display.flip()  # Ekranı güncelle

            pygame.time.delay(100)  # 100 ms bekle (animasyon hızını kontrol eder)

        pygame.time.delay(500)  # Son kareyi ekranda biraz daha uzun tut
    else:
        print(f"'{kelime}' için animasyon bulunamadı.")

# Ana akış
metin = sesli_komut_al()
if metin:
    kelimeler_in_metin = metin.split()
    if kelimeler_in_metin:  # Eğer metinde kelimeler varsa
        hedef_kelime = kelimeler_in_metin[0]  # İlk kelimeyi alıyoruz
        print(f"Söylediğiniz kelime: {hedef_kelime}")
        animasyon_oynat(hedef_kelime)  # Bu kelime için animasyonu oynat
    else:
        print("Hiç kelime algılanamadı.")

# Pygame ekranını kapat
pygame.quit()
```

```python
import speech_recognition as sr
import os
import cv2

# İşaret dili videolarının bulunduğu klasör
VIDEO_FOLDER = "İşaretDiliVideo"
OUTPUT_VIDEO = "output.mp4"  # Kaydedilecek dosyanın adı

# Konuşmayı Algılama Fonksiyonu
def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Konuşun... (Maksimum 10 saniye)")
        recognizer.adjust_for_ambient_noise(source, duration=1)  # Gürültü ayarı

        try:
            audio = recognizer.listen(source, timeout=10, phrase_time_limit=10)  # Maksimum 10 saniye dinle
            text = recognizer.recognize_google(audio, language="tr-TR")
            print(f"Saptanan Cümle: {text}")
            return text.lower()  # Küçük harfe çevirerek dosya adlarıyla eşleşmesini sağlıyoruz
        except sr.UnknownValueError:
            print("Ses algılanamadı.")
            return None
        except sr.RequestError:
            print("Google API hatası.")
            return None
        except sr.WaitTimeoutError:
            print("Zaman aşımı! 10 saniye içinde konuşma algılanmadı.")
            return None

# Cümledeki kelimeleri videolarla eşleştirme
def find_video_files(sentence):
    words = sentence.split()  # Cümleyi kelimelere ayır
    video_paths = []
    
    # Klasördeki tüm dosyaları küçük harfe çevirerek al
    files = [file.lower() for file in os.listdir(VIDEO_FOLDER) if file.endswith((".mp4", ".avi", ".mov"))]

    print(f"Klasördeki Videolar: {files}")

    for word in words:
        print(f"\nKelime kontrol ediliyor: {word}")
        for file in files:
            # Eğer kelime dosya adında geçiyorsa ekle
            if word in file:
                full_path = os.path.join(VIDEO_FOLDER, file)
                if full_path not in video_paths:  # Aynı videoyu tekrar eklememek için kontrol
                    print(f"✔ Video eklendi: {full_path}")
                    video_paths.append(full_path)

    print(f"\nBulunan Videolar: {video_paths}")  # Kontrol için ekrana yazdır
    return video_paths

# Videoları sırayla oynatma ve kaydetme
def play_and_save_videos(video_paths):
    if not video_paths:
        print("Eşleşen kelimeler için video bulunamadı.")
        return

    # İlk videonun çözünürlüğünü ve FPS'ini belirleme
    first_video = cv2.VideoCapture(video_paths[0])
    if not first_video.isOpened():
        print("Hata: İlk video açılamadı!")
        return
    
    frame_width = int(first_video.get(3))
    frame_height = int(first_video.get(4))
    fps = int(first_video.get(cv2.CAP_PROP_FPS)) if first_video.get(cv2.CAP_PROP_FPS) > 0 else 30
    first_video.release()

    # Video kaydedici oluştur
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # MP4 formatında kaydetmek için
    out = cv2.VideoWriter(OUTPUT_VIDEO, fourcc, fps, (frame_width, frame_height))

    for path in video_paths:
        cap = cv2.VideoCapture(path)

        if not cap.isOpened():
            print(f"Hata: {path} açılamadı!")
            continue

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break  # Video bittiğinde döngüden çık
            
            cv2.imshow("İşaret Dili Çeviri", frame)
            out.write(frame)  # Kareyi kaydet

            # 'q' tuşuna basılırsa çık
            if cv2.waitKey(30) & 0xFF == ord('q'):
                cap.release()
                out.release()
                cv2.destroyAllWindows()
                return
        
        cap.release()  # Video bittiğinde serbest bırak

    out.release()  # Kaydediciyi kapat
    cv2.destroyAllWindows()  # Tüm videolar bitince kapat
    print(f"Kaydedilen video: {OUTPUT_VIDEO}")

# Ana Fonksiyon
def main():
    detected_sentence = recognize_speech()
    if detected_sentence:
        video_paths = find_video_files(detected_sentence)
        if video_paths:
            print(f"Oynatılacak Videolar: {video_paths}")
            play_and_save_videos(video_paths)  # Videoları sırayla oynat ve kaydet
        else:
            print("Cümledeki kelimeler için video bulunamadı.")
    else:
        print("Cümle algılanamadı.")

if __name__ == "__main__":
    main()
```

```python
import os
import json
import cv2

JSON_FOLDER = "jsonDosyasi"  # JSON dosyalarının bulunduğu klasör
VIDEO_FRAMES_FOLDER = "frames"  # Frame görsellerinin bulunduğu klasör

def get_frame_paths_for_words(words):
    frame_paths = []
    
    for word in words:
        json_file_path = os.path.join(JSON_FOLDER, f"{word}.json")
        
        if os.path.exists(json_file_path):
            with open(json_file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
            
            if "frames" in data:
                for frame in data["frames"]:
                    if isinstance(frame, dict) and "image_path" in frame:
                        frame_path = os.path.join(frame["image_path"]).replace("\\", "/")
                        
                        if os.path.exists(frame_path):
                            frame_paths.append(frame_path)
        else:
            print(f"Uyarı: {word}.json dosyası bulunamadı.")
    
    return frame_paths

def play_frames(frame_paths):
    for frame_path in frame_paths:
        frame = cv2.imread(frame_path)
        if frame is not None:
            cv2.imshow("Frame", frame)
            cv2.waitKey(100)  # Her kareyi 100ms gösterir
        else:
            print(f"Hata: {frame_path} yüklenemedi.")
    cv2.destroyAllWindows()

def main():
    detected_sentence = input("Konuştuğunuz cümleyi girin: ")  # Kullanıcıdan cümle al
    if detected_sentence:
        words = detected_sentence.split()  # Cümledeki kelimeleri ayır
        frame_paths = get_frame_paths_for_words(words)  # JSON'dan frame yollarını al
        play_frames(frame_paths)  # Frame'leri sırayla göster

if __name__ == "__main__":
    main()
```
