```python
import cv2
import os
import json

def extract_frames(video_path, output_folder, output_json, fps=30):
    """Videoyu frame'lere ayırır ve JSON dosyasına kaydeder."""
    
    # Klasör oluştur
    os.makedirs(output_folder, exist_ok=True)
    
    # Video aç
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    frame_data = {"frames": []}
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Video bitti

        # Frame kaydet
        frame_filename = f"frame_{frame_count:04d}.jpg"
        frame_path = os.path.join(output_folder, frame_filename)
        cv2.imwrite(frame_path, frame)

        # Zaman hesaplama
        timestamp = frame_count / fps

        # JSON içine frame bilgisi ekleme
        frame_info = {
            "frame_id": frame_count + 1,
            "image_path": frame_path,
            "timestamp": round(timestamp, 3),
            "keypoints": []  # İskelet verisi varsa buraya eklenebilir
        }
        frame_data["frames"].append(frame_info)
        
        frame_count += 1

    cap.release()

    # JSON dosyasını kaydet
    with open(output_json, "w", encoding="utf-8") as f:
        json.dump(frame_data, f, indent=4, ensure_ascii=False)

    print(f"{frame_count} frame çıkarıldı ve JSON kaydedildi: {output_json}")

# Kullanım
video_path = "İşaretDiliVideo\Ben.mp4"  # İşlenecek video
output_folder = "frames"  # Frame'lerin kaydedileceği klasör
output_json = "frames_data_ben.json"  # Çıktı JSON dosyası

extract_frames(video_path, output_folder, output_json)
```
